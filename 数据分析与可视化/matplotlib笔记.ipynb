| 图表类型 | 适用数据 | 常见用途 |
| :--- | :--- | :--- |
| 折线图 line | 连续变量 | 时间趋势、变化轨迹 |
| 条形图 bar | 分类数据 | 类别比较 |
| 饼图 pie | 构成比例 | 占比展示 |
| 散点图 scatter | 数值对数值 | 相关关系、聚类 |
| 箱线图 box | 数值型 | 数据分布、异常值 |
| 热力图 heatmap | 表格型 | 关联性、密度分布 |

不同开发环境下显示图形  
- 在一个脚本文件中使用Matplotlib，那么显示图形的时候必须使用plt.show()。

```python
import matplotlib.pyplot as plt

# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题
```

## 1. 两种画图接口

### 1.1 状态接口  
*便捷的MATLAB风格的有状态的接口*

```python
# 折线图
month = ['一月', '二月', '三月', '四月', '五月', '六月']
sales = [100, 120, 150, 180, 200, 250]
# 创建折线图,并设置大小
plt.figure(figsize=(10, 6))
# 绘制折线图，marker表示数据点的形状，markerfacecolor表示数据点填充颜色，
# markersize表示数据点大小, label表示图例名称
plt.plot(month, sales, color='purple', linewidth=2, marker='o', 
         markerfacecolor='blue', markersize=8,
         label='产品A')
# 添加标题和标签，pad表示标题与图形的距离
plt.title('2025年上半年销售额折线图', fontsize=20, color='red', pad=20)
plt.xlabel('月份', fontsize=14)
plt.ylabel('销售额（单位：万元）', fontsize=14)
# 添加图例，loc表示位置
plt.legend(loc = 'upper left')
# 显示网格，linestyle表示线型，alpha表示透明度，axis表示显示哪个轴的网格
# true表示显示网格
plt.grid(True, linestyle='--', alpha=0.7, color='gray')
# 自定义刻度字体大小，rotation表示旋转角度
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12,rotation=45)
# 自定义y轴范围
plt.ylim(80, 270)
# 在每个数据点上方添加数值标签,ha表示水平对齐方式，va表示垂直对齐方式
for xi, yi in zip(month, sales):
    plt.text(xi, yi + 5, str(yi), ha='center', va='bottom', fontsize=10, color='black')
# 自动优化排版
plt.tight_layout()
# 显示图形
plt.show()
```

```python
# 散点图（展示两个数值变量的关系）
import numpy as np

x_data = np.random.randint(1,100,200)
y_data = 2*x_data+np.random.uniform(0,10,200)
plt.figure(figsize=(10, 6))
# 绘制散点图，s表示点的大小，alpha表示透明度，edgecolors表示点的边框颜色
plt.scatter(x_data, y_data, color='green', s=100, alpha=0.6, edgecolors='black', label='数据点')
plt.title('散点图示例', fontsize=20, color='blue', pad=20)
plt.xlabel('自变量X', fontsize=14)  
plt.ylabel('因变量Y', fontsize=14)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7, color='gray')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 柱状图（对比不同类别的数据大小）

# 类别和对应数值
subjects = ['语文', '数学', '英语', '物理', '化学', '生物']
scores = [85, 90, 78, 88, 92, 80]
# 创建条形图,并设置大小
# edgecolor表示边框颜色，width表示条形宽度，label表示图例名称
plt.figure(figsize=(10, 6))
plt.bar(subjects, scores, color='skyblue', edgecolor='black',
        width = 0.6, label='2025年上半年成绩')
# 添加标题和标签
plt.title('2025年上半年各科成绩图', fontsize=20, color='red', pad=20)
plt.xlabel('科目', fontsize=14)
plt.ylabel('成绩（分数）', fontsize=14)
# 添加图例
plt.legend(loc = 'upper right')
# 显示网格
plt.grid(True, linestyle='--', alpha=0.7, color='gray', axis='y')
# 设置y轴范围
plt.ylim(0, 100)
# 自定义刻度字体大小
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
#  为每个数值添加标签,enumerate函数用于获取索引和值
for i, score in enumerate(scores):
    # print(i, score)
    plt.text(i, score +2, str(score), ha = 'center', va = 'bottom', fontsize=12, color='black')
plt.tight_layout()
plt.show()
```

```python
# 条形图
# 类别和对应数值
categories = ['A类', 'B类', 'C类', 'D类', 'E类']
values = [40, 70, 30, 85, 60]
# 创建水平条形图,并设置大小
plt.figure(figsize=(10, 6))
plt.barh(categories, values, color='lightgreen', edgecolor='black',
         height=0.6, label='2025年各类别数值')
# 添加标题和标签
plt.title('2025年各类别数值水平条形图', fontsize=20, color='red', pad=20)
plt.xlabel('数值', fontsize=14)
plt.ylabel('类别', fontsize=14)
# 添加图例
plt.legend(loc = 'lower right')
# 显示网格
plt.grid(True, linestyle='--', alpha=0.7, color='gray', axis='x')
# 设置x轴范围
plt.xlim(0, 100)
# 自定义刻度字体大小
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
# 为每个数值添加标签
for i, value in enumerate(values):
    # print(i, value)
    plt.text(value + 2, i, str(value), ha='left', va='center', fontsize=12, color='black')
plt.tight_layout()
plt.show()
```

```python
# 饼图（显示各部分占比，时间分配，市场份额）
# 最好不要超过6类。3~5类最佳
# 各部分标签和对应数值
labels = ['学习', '工作', '娱乐', '锻炼', '休息']
time_spent = [40, 30, 15, 10, 5]
# 创建饼图,并设置大小
plt.figure(figsize=(6, 6))
# autopct表示百分比格式，%%转意显示出来，startangle表示起始角度，colors表示各部分颜色
plt.pie(time_spent, labels=labels, autopct='%1.1f%%',
        startangle=140, 
        colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'violet'])
# 添加标题
plt.title('每日时间分配饼图', fontsize=20, color='red', pad=20)
plt.tight_layout()
plt.show()
```

```python
# 环形图
labels = ['学习', '工作', '娱乐', '锻炼', '休息']
time_spent = [40, 30, 15, 10, 5]
plt.figure(figsize=(6, 6))
# autopct表示百分比格式，%%转意显示出来，startangle表示起始角度，colors表示各部分颜色
plt.pie(time_spent, labels=labels, autopct='%1.1f%%',
        startangle=140, 
        shadow=True,
        wedgeprops={'width':0.6},  # 设置环形宽度
        pctdistance=0.75,          # 设置百分比标签距离圆心的距离
        colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'violet'])
# 添加标题
plt.title('每日时间分配环形图', fontsize=20, color='red', pad=20)
plt.text(0, 0, '时间分配', ha='center', va='center', fontsize=14, color='black')
plt.tight_layout()
plt.show()
```

```python
# 爆炸式饼图
labels = ['学习', '工作', '娱乐', '锻炼', '休息']
time_spent = [40, 30, 15, 10, 5]
explode = (0.1, 0, 0, 0, 0)  # 仅“学习”部分突出显示
# 创建饼图,并设置大小
plt.figure(figsize=(6, 6))
# autopct表示百分比格式，%%转意显示出来，startangle表示起始角度，colors表示各部分颜色
plt.pie(time_spent, labels=labels, autopct='%1.1f%%',
        startangle=0, explode=explode,# 设置突出部分
        shadow=True,               # 添加阴影效果
        colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'violet'])
# 添加标题
plt.title('每日时间分配饼图', fontsize=20, color='red', pad=20)
plt.tight_layout()
plt.show()
```

```python
# 模拟 3 门课的成绩
data = {
    '语文': [82, 85, 88, 70, 90, 76, 84, 83, 95],
    '数学': [75, 80, 79, 93, 88, 82, 87, 89, 92],
    '英语': [70, 72, 68, 65, 78, 80, 85, 90, 95]
}

plt.figure(figsize=(8, 6))
plt.boxplot(data.values(), labels=data.keys())

plt.title("各科成绩分布（箱线图）")
plt.ylabel("分数")
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()
```

- 中位数：盒子中间的线  
- 上/下四分位数：盒子上下边缘  
- 离群值：落在“胡须”外的点  
1. 数学成绩  
中位数最高（约88分），且箱体较短 → 学生成绩集中且整体较好。  
无异常值 → 无极端高分或低分。  
2. 语文成绩  
箱体较长 → 成绩分布较分散（从70分到95分）。  
上方有一个异常值（95分）→ 可能存在个别高分学生。  
3. 英语成绩  
中位数最低（约78分），但箱须向上延伸较长 → 部分学生成绩较高（90+分）。  
下方无异常值 → 无极端低分。

```python
# 多个图标绘制
# plt.subplot()  子图
plt.figure(figsize=(12, 8))
# 第一个子图：折线图
plt.subplot(2, 2, 1)  # 2行2列的第1个子图
month = ['一月', '二月', '三月', '四月', '五月', '六月']
sales = [100, 120, 150, 180, 200, 250]
plt.plot(month, sales, color='purple', linewidth=2, marker='o', 
         markerfacecolor='blue', markersize=8,
         label='产品A')
plt.title('2025年上半年销售额折线图', fontsize=16, color='red', pad=10)
plt.xlabel('月份', fontsize=12)
plt.ylabel('销售额（单位：万元）', fontsize=12)
plt.legend(loc = 'upper left')
plt.grid(True, linestyle='--', alpha=0.7, color='gray')
# 第二个子图：散点图
plt.subplot(2, 2, 2)  # 2行2列的第2个子图
x_data = np.random.randint(1,100,200)
y_data = 2*x_data+np.random.uniform(0,10,200)
plt.scatter(x_data, y_data, color='green', s=100, alpha=0.6, edgecolors='black', label='数据点')
plt.title('散点图示例', fontsize=16, color='blue', pad=10)
plt.xlabel('自变量X', fontsize=12)
plt.ylabel('因变量Y', fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7, color='gray')
# 第三个子图：柱状图
plt.subplot(2, 2, 3)  # 2行2列的第3个子图
subjects = ['语文', '数学', '英语', '物理', '化学', '生物']
scores = [85, 90, 78, 88, 92, 80]
plt.bar(subjects, scores, color='skyblue', edgecolor='black',
        width = 0.6, label='2025年上半年成绩')
plt.title('2025年上半年各科成绩图', fontsize=16, color='red', pad=10)
plt.xlabel('科目', fontsize=12)
plt.ylabel('成绩（分数）', fontsize=12)
plt.legend(loc = 'upper right')
plt.grid(True, linestyle='--', alpha=0.7, color='gray', axis='y')
# 第四个子图：饼图
plt.subplot(2, 2, 4)  # 2行2列的第4个子图
labels = ['学习', '工作', '娱乐', '锻炼', '休息']
time_spent = [40, 30, 15, 10, 5]
plt.pie(time_spent, labels=labels, autopct='%1.1f%%',
        startangle=140, 
        colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'violet'])
plt.title('每日时间分配饼图', fontsize=16, color='red', pad=10)
plt.tight_layout()
plt.show()
```
